```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
import pandas as pd

# Load the dataset
file_path = 'cleaned_final_data.csv'
data = pd.read_csv(file_path)

# Split the data into training and testing datasets
train_data = data.sample(frac=0.8, random_state=42)  # 80% for training
test_data = data.drop(train_data.index)  # remaining 20% for testing

# Specify the label
label = 'flooded'

# Create a TabularPredictor object
predictor = TabularPredictor(label=label).fit(train_data)

```

```python

predictor = TabularPredictor.load("\AutogluonModels\ag-20240429_230251")
```

```python
predictor.leaderboard(test_data)
```

```python
# Evaluate the model on the test data
performance = predictor.evaluate(test_data)
```

```python
# Save the best performing model
best_model = predictor.get_model_best()
predictor.save_space('/best_model')

print("Model evaluation:", performance)
print("Best model:", best_model)
```
